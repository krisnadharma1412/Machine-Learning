```python
from scipy.stats import bernoulli
data_bern=bernoulli.rvs(size=10000,p=0.6)
```

```python
# Plot a histogram and kernel density estimate
sns.distplot(data_bern,kde=False,color="r")
plt.xlabel('Bernoulli Distribution')
plt.ylabel('Frequency')
```

```python
# import uniform distribution
from scipy.stats import uniform
```

```python
# random numbers from uniform distribution
n=10000
start=10
width=20
data_uniform=uniform.rvs(size=n,loc=start,scale=width)
```

```python
# Plot a histogram and kernel density estimate
sns.distplot(data_uniform,bins=100,color="r")
plt.xlabel('Uniform Distribution')
plt.ylabel('Frequency')
```

```python
from scipy.stats import norm
# generate random numbers from N(0,1)
data_normal=norm.rvs(size=10000,loc=0,scale=1)
```

```python
# Plot a histogram and kernel density estimate
sns.distplot (data_normal, color="g")
plt.xlabel('Normal Distribution')
plt.ylabel('Frequency')
```

```python
from scipy.stats import expon
data_expon=expon.rvs(scale=1,loc=0,size=1000)
```

```python
# Plot a histogram and kernel density estimate
sns.distplot(data_expon,color="y")
plt.xlabel('Exponential Distribution')
plt.ylabel('Frequency')
```

```python
from scipy.stats import beta
a,b=2.30984964515,0.62687954301
data_beta=beta.rvs(a,b,size=1000)
```

```python
# Plot a histogram and kernel density estimate
sns.distplot(data_beta,color="g")
plt.xlabel('Beta Distribution')
plt.ylabel('Frequency')
```
